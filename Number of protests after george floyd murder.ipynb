```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('2019-01-22-2022-01-27-North_America-United_States.csv')
```

```python
year2020 = data[data.year == 2020]
```

```python
may25 = year2020[(year2020["event_date"] == "25-May-20") & (year2020["event_type"] == "Protests")]
```

```python
may26 = year2020[(year2020["event_date"] == "26-May-20") & (year2020["event_type"] == "Protests")]
```

```python
may27 = year2020[(year2020["event_date"] == "27-May-20") & (year2020["event_type"] == "Protests")]
```

```python
may28 = year2020[(year2020["event_date"] == "28-May-20") & (year2020["event_type"] == "Protests")]
```

```python
may29 = year2020[(year2020["event_date"] == "29-May-20") & (year2020["event_type"] == "Protests")]
```

```python
may30 = year2020[(year2020["event_date"] == "30-May-20") & (year2020["event_type"] == "Protests")]
```

```python
may31 = year2020[(year2020["event_date"] == "31-May-20") & (year2020["event_type"] == "Protests")]
```

```python
x = [25,26,27,28,29,30,31]
y = [len(may25),len(may26),len(may27),len(may28),len(may29),len(may30),len(may31)]
plt.plot(x, y)
plt.title("Number of protests after May 25th, 2020")
plt.xlabel("Date")
plt.ylabel("Number of Protests")
plt.grid(True)
##plt.legend(["This is y"])
plt.show()
```
